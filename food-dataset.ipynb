```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python


file_path = '/kaggle/input/indian-food-101/indian_food.csv'
columns_name = ['name','ingredients','diet','prep_time','cook_time','flavor_profile','course','state','region']
df = pd.read_csv(file_path,header=None,names=columns_name)
display(df.head())
```

```python
df.replace("?", pd.NA, inplace=True)
df.dropna(inplace=True)
print(df.head())
```

```python
[col for col in df.colums if df[col].isnull().sum() > 0]
```

```python
x_train , x_test , y_train , y_test = train_test_split (df.drop(labels = ['prep_time'], axis=1), df['prep_time'],test_size = 0.3, random_state = 0)

constant_feature = [ features for features in x_train.columns if x_train[features].nunique() == 1]
len(constant_feature)
```

```python
quasi_constant_feature = []

for features in x_train.columns:
    predominant = (x_train[features].value_counts() / float(len(x_train))).sort_values(ascending=False).values[0]

    if predominant > 0.999:
        quasi_constant_feature.append(feature)

len(quasi_constant_feature)
```

```python
duplicated_feature = []

for i in range(len(x_train.columns)):
    col_1 = x_train.columns[i]

    for col_2 in x_train.columns[i + 1:]:
        if x_train[col_1].equals(x_train[col_2]):
            duplicated_feature.append(col_2)

duplicated_features = set(duplicated_feature)
print(len(duplicated_features))
```
